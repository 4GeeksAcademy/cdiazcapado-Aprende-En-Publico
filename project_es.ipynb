#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
calif_estu = [8.5, 9.2, 7.8, 8.9, 10]
total_sum = sum(calif_estu)
cant_total = len(calif_estu)
prome = total_sum/cant_total
print(f"el promedio de la calificaciones es",{prome} )
```

- Encuentra la calificación más alta y la más baja.

```python
# Código

mas_alta = max(calif_estu)
print(f" la calificacion mas alta es", {mas_alta})
mas_baja = min(calif_estu)
print(f" la calificacion mas baja es", {mas_baja})
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
ds_arroyomolinos = ds.loc[ds['level5'] == 'Arroyomolinos (Madrid)']
print(ds_arroyomolinos)
print(f"La media de precios en Arroyomolinos, Madrid es: {ds_arroyomolinos['price'].mean()}")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios

plt.hist(ds_arroyomolinos["price"], bins=137)
plt.title ("precios Arroyomolinos Madrid")
plt.xlabel("precios")
plt.ylabel("unidades viviendas")
plt.show()
# en el sector de Arroyomolinos se observan viviendas que oxcilan entre diferentes precios, y hay viviendas desde los 160000.
```
